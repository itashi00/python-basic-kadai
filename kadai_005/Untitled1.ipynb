```python
import matplotlib.pyplot as plt
import numpy as np

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# プロット
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

plt.show()
```

```python
# Step1: matplotlib.pyplot と numpy をインポート
import matplotlib.pyplot as plt
import numpy as np

# Step2: 6つのグラフ領域を作成 (2行3列)、レイアウト自動調整有効
fig, axes = plt.subplots(2, 3, tight_layout=True)

# Step3: 各領域にグラフを描画

# 折れ線グラフ（axes[0,0]）
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('折れ線グラフ')

# sin関数グラフ（axes[0,1]）
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
axes[0, 1].plot(x, y1)
axes[0, 1].set_title('sin関数')

# ヒストグラム（axes[0,2]）
data = np.random.randn(1000)
axes[0, 2].hist(data, bins=30, color='orange', alpha=0.7)
axes[0, 2].set_title('ヒストグラム')

# 散布図（axes[1,0]）
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
axes[1, 0].scatter(x_scatter, y_scatter, color='green')
axes[1, 0].set_title('散布図')

# その他2つの領域は空白のまま

# Step4: グラフを表示
plt.show()
```
