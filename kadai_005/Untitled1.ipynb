```python
import matplotlib.pyplot as plt
import numpy as np

# 
# 010100
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# 
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

plt.show()
```

```python
# Step1: matplotlib.pyplot  numpy 
import matplotlib.pyplot as plt
import numpy as np

# Step2: 6 (23)
fig, axes = plt.subplots(2, 3, tight_layout=True)

# Step3: 

# axes[0,0]
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('')

# sinaxes[0,1]
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
axes[0, 1].plot(x, y1)
axes[0, 1].set_title('sin')

# axes[0,2]
data = np.random.randn(1000)
axes[0, 2].hist(data, bins=30, color='orange', alpha=0.7)
axes[0, 2].set_title('')

# axes[1,0]
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
axes[1, 0].scatter(x_scatter, y_scatter, color='green')
axes[1, 0].set_title('')

# 2

# Step4: 
plt.show()
```
